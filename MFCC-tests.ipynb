```python
import librosa
import matplotlib.pyplot as plt
import scipy
```

```python
data, rate = librosa.load("oboe-g4-PB-loop.wav")
```

```python
def show_data(data):
    plt.figure()
    plt.plot(data)
    plt.show()
    
show_data(data)
```

```python
data2, rate2 = librosa.load("1st-violins-stc-rr2-g4-PB.wav")
show_data(data2)
```

```python
mfccs = librosa.feature.mfcc(data, rate)
mfccs[:,0]
```

```python
mfccs2 = librosa.feature.mfcc(data2, rate2)
mfccs2[:,0]
```

```python
scipy.spatial.distance.cosine(mfccs[:,2], mfccs2[:,2])
```
